```python
#python program to cody odd lines of one file to other
file1=input("Enter the source file")
file2=input("Enter the destination file")
f1=open(file1,"r")
f2=open(file2,"w")
i=0
for line in f1.readlines():
    i=i+1
    if i%2==0:
        continue
    f2.write(line)
f1.close()
f2.close()
print("The odd lines of",file1,"is transfered to",file2)
```

```python
#write a python program to read each row from a given csv file and print a list of strings
import csv
fname=input("Enter the csv file name:")
f1=open(fname,"r")
reader=csv.reader(f1)
for row in reader:
    print(row)
f1.close()
```

```python
#write a python program to read specific columns of a given CSV file and print the contents of the columns
import csv
fname=input("Enter the CSV file name:")
with open(fname,"r") as f1:
    reader=csv.DictReader(f1)
    print("\nDepartment Details:")
    print("------------------------")
    for row in reader:
        print("DeptId:",row['DeptId'],"\tDeptName",row['DeptName'])
```

```python
#write a python program to write a Python dictonary to a new csv file. after writing the CSV file read the CSV file and display the content
import csv
data=[
    {"Name":"resmi","Age":22,"Marks":85},
    {"Name":"arun","Age":21,"Marks":75},
    {"Name":"meera","Age":22,"Marks":80}
]
with open("student_dict.csv","w",newline='') as file:
    fieldnames=["Name","Age","Marks"]
    writer=csv.DictWriter(file,fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(data)
print("Dictionary data written successfully")
with open("student_dict.csv","r") as file:
    reader=csv.reader(file)
    for row in reader:
        print(row)
    
```
