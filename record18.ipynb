```python
#CREATE A CLASS RECTANGLE WITH PRIVATE ATTRIBUTES LENGTH AND WIDTH.OVERLOAD "<" OPERATOR TO COMPARE THE AREA OF TWO RECTANGLE
class rectangle:
    def __init__(self,length,width):
        self.length=length
        self.width=width
        self.area=length*width
    def display(self):
        return self.area
    def __lt__(self,other):
        if self.area<other.area:
            print("Rectangle 2 is greater")
        else:
            print("Rectangle 1 is greater")
        
l1=int(input("Enter the length of recatngle1"))
w1=int(input("Enter the width of recatngle1"))
l2=int(input("Enter the length of recatngle2"))
w2=int(input("Enter the width of recatngle2"))
r1=rectangle(l1,w1)
r2=rectangle(l2,w2)
print("Area of rectangle 1 is :",r1.display())
print("Area of rectangle 2 is :",r2.display())
r1<r2
```

```python
#create a class time with private attributes hour,minute and second.overload "+" operator to find the sum of 2 time
class time:
    def __init__(self,hour,minute,second):
        self.hour=hour
        self.minute=minute
        self.second=second
    def display(self):
        print("Time=",self.hour,":",self.minute,":",self.second)
    def __add__(self,other):
        h=self.hour+other.hour
        m=self.minute+other.minute
        s=self.second+other.second
        if s>=60:
            m+=s//60
            s=s%60
        if m>=60:
            h+=m//60
            m=m%60
        return time(h,m,s)
print("Enter first time:")
h1 = int(input("Hour: "))
m1 = int(input("Minute: "))
s1 = int(input("Second: "))
print("\nEnter second time:")
h2 = int(input("Hour: "))
m2 = int(input("Minute: "))
s2 = int(input("Second: "))
t1 = time(h1, m1, s1)
t2 = time(h2, m2, s2)
t3 = t1 + t2   
print("First time:")
t1.display()
print("Second time:")
t2.display()
print("Sum of times:")
t3.display()
```
